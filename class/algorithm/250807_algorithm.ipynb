## ⭐ 부분집합
유한 개의 정수로 이루어진 집합이 있을 때, 이 집합의 부분집합 중에서 조건에 맞는 경우를 찾는 문제

- 예시 : 그 집합의 원소를 모두 더한 값이 0이 되는 경우가 있는지를 알아내는 문제

1. 부분집합의 원소를 표현하는 방법

```python
def print_subset(bit):
    for i in range(4):
        if bit[i]: # bit[i]가 0이 아니면
            print(arr[i], ' ')
    print(bit)

arr = [7, 5, 8, 1]

bit = [0, 0, 0, 0]

for i in range(2):
    bit[0] = i
    for j in range(2):
        bit[1] = j
        for k in range(2):
            bit[2] = k
            for l in range(2):
                bit[3] = l
                print_subset(bit)
```

##### 비트 연산자
- & : 비트 단위로 AND 연산
- | : 비트 단위로 OR 연산
- '<<' : 피연산자의 비트 열을 왼쪽으로 이동
- '>>' : 피연산자의 비트 열을 오른쪽으로 이동

##### << 연산자 활용
- 1<<n : 2^n 즉, 원소가 n개일 경우의 모든 부분집합의 수를 의미

##### & 연산자 활용
- i & (i<<j) : i의 j번째 비트가 1인지 아닌지를 검사

##### 특별한 정렬

1. 보통의 정렬은 오름차순이나 내림차순으로 이루어지지만, 이번에는 특별한 정렬을 하려고 한다.  
2. N개의 정수가 주어지면 가장 큰 수, 가장 작은 수, 2번째 큰 수, 2번째 작은 수 식으로 큰 수와 작은 수를 번갈아 정렬하는 방법이다.  
3. 예를 들어 1부터 10까지 10개의 숫자가 주어지면 다음과 같이 정렬한다.  
 
10 1 9 2 8 3 7 4 6 5
 
4. 주어진 숫자에 대해 특별한 정렬을 한 결과를 10개까지 출력하시오

[입력]  
1. 첫 줄에 테스트 케이스 개수 T가 주어진다.  1<=T<=50
2. 다음 줄에 정수의 개수 N이 주어지고 다음 줄에 N개의 정수 ai가 주어진다. 10<=N<=100, 1<=ai<=100  

[출력]  
1. 각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, ❗특별히 정렬된 숫자를 10개까지 출력한다.  

```python
import sys
sys.stdin = open('input.txt')

# 함수 1 : 최대값 구하기 
def find_max_value(random_array):
    max_value = random_array[0]
    for i in range(len(random_array)):
        if max_value < random_array[i]:
            max_value = random_array[i]
    return max_value

# 함수 2 : 최소값 구하기
def find_min_value(random_array):
    min_value = random_array[0]
    for i in range(len(random_array)):
        if min_value > random_array[i]:
            min_value = random_array[i]
    return min_value


# 테스트 케이스 입력 받기
T = int(input())

# 테스트 케이스만큼 수행
for time in range(1, T+1):
    # 정수의 개수 = 배열의 길이 받기
    N = int(input())
    # 배열의 값 받기
    arr = list(map(int, input().split()))

    # 정렬된 값을 받을 새로운 배열 할당
    new_arr = []

    # 함수를 수행하되, 배열의 길이의 절반만 수행하기
    # 최대값과 최소값을 함께 정렬하기 때문에 배열의 길이에서 절반만 수행해야 함
    if N % 2 == 0 :
        for i in range(5):
            # 최대값, 최소값 함수 수행
            max_value = find_max_value(arr)
            min_value = find_min_value(arr)

            # 최대값 먼저 append한 이후에 최소값 append 함
            new_arr.append(max_value)
            new_arr.append(min_value)

            # 정렬된 값은 기존 배열에서 삭제함
            arr.remove(max_value)
            arr.remove(min_value)
        
        print(f'#{time}', end=' ')
        for i in new_arr:
            print(i, end=' ')
        print()    
```

6. 책의 전체 쪽수와 두 사람이 찾을 쪽 번호가 주어졌을 때, 이진 탐색 게임에서 이긴 사람이 누구인지 알아내 출력하시오. 비긴 경우는 0을 출력한다.

[입력]  
 
1. 첫 줄에 테스트 케이스 개수 T가 주어진다.  1<=T<=50  
2. 테스트 케이스 별로 책의 전체 쪽 수 P, A, B가 찾을 쪽 번호 Pa, Pb가 차례로 주어진다. 1<= P, Pa, Pb <=1000  
 
[출력]  
1. 각 줄마다 "#T" (T는 테스트 케이스 번호)를 출력한 뒤, A, B, 0 중 하나를 출력한다.

```python
import sys
sys.stdin = open('input2.txt')

# 북 페이지 찾는 함수 작성
def search_book_page(total_page, find_page):
    # 이진탐색을 위한 start, end 변수 할당
    # start를 1로 주어야 함 (주어진 조건)
    start_page = 1
    end_page = total_page

    # 이진탐색 횟수를 세는 count 변수 할당
    count = 0

    # 이진탐색 시작
    while start_page <= end_page:
        # end_page가 홀수로 끝날 수도 있으니, 해당 값까지 반영해야 함
        half = int((start_page + end_page) / 2)
        if half == find_page:
            return count
            break
        elif half > find_page:
            end_page = half
            count += 1
        elif half < find_page:
            start_page = half 
            count += 1
    return count

T = int(input())

for time in range(1, T+1):
    P, A, B = map(int,input().split())

    result_A = search_book_page(P, A)
    result_B = search_book_page(P, B)

    # return count 변수가 작은 사람을 출력함
    if result_A > result_B:
        print(f'#{time} B')
    elif result_A == result_B:
        print(f'#{time} 0')
    elif result_A < result_B:
        print(f'#{time} A')
```

```python
import sys
sys.stdin = open('input4.txt')

ladder_arr = [list(map(int, input().split())) for _ in range(100)]

start_point = 0

# y : 도착지점 값 할당
for z in range(100):
    if ladder_arr[99][z] == 2:
        start_point = z

def find_one(arr, x, y):
    left_right_row = [-1, 1, 0]
    up_column = [0, 0, -1]

    for idx in range(3):
        new_r = x + up_column[idx]
        new_c = y + left_right_row[idx]
        if new_r < 0 or new_r >= 100 or new_c < 0 or new_c >= 100:
            break

        if arr[new_r][new_c] == 1:
            # 좌우로 움직일 수 있으면,
            if idx < 2:
                direction = left_right_row[idx]
                i = 0
                while arr[x][y] > 0 and  x < 100 and y < 100 and x >= 0 and y >= 0:
                    y = y + (direction * i)
                    i += 1
                print(x, y)
                return x-1, y
            # 좌우로 움직일 길이 없으면,
            else:
                x -= 1
                return x, y
        else:
            continue
        
r = 99
c = start_point
while r != 0:
    result = find_one(ladder_arr, r, c) # r이 row, c가 column
    print(result)


    r = result[0]
    c = result[1]
```

```python
def find_subsets_bitmask(arr):
    n = len(arr)
    # 총 부분집합의 개수 = 2^n
    for i in range(1 << n):  # 1 << n == 2**n
        subset = []
        for j in range(n):
            # i의 j번째 비트가 1이면 해당 원소를 포함
            if (i & (1 << j)):
                subset.append(arr[j])
        print(subset)

# 예시: {1, 2, 3}의 부분집합
my_set = [1, 2, 3]
find_subsets_bitmask(my_set)
```
